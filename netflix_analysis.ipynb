```python
df['director'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown',inplace=True)
df['rating'].fillna(df['rating'].mode()[0],inplace=True)
```

```python
sns.heatmap(df.isnull())
plt.show
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
sns.heatmap(df.isnull())
plt.show
```

```python
df.dropna(inplace=True)
```

```python
sns.heatmap(df.isnull())
plt.show()
```

```python
df['type'].value_counts()
```

```python
sns.countplot(x='type',data=df)
plt.title("MOVIES vs TV SHOWS on Netflix")
plt.show()
```

```python
df['country'].value_counts().head(10)
```

```python
plt.figure(figsize=(15,10))
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 countries on Netflix")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show
```

```python
df['rating'].value_counts()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index)
plt.title("Netflix Content Ratings")
plt.show()
```

```python
df['release_year'].value_counts().head(10)
```

```python
plt.figure(figsize=(12,5))

df['release_year'].value_counts().sort_index().plot()

plt.title("Netflix Content Released by Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.show()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df['year_added'] = df['date_added'].dt.year
```

```python
df[['title','year_added']].head()
```

```python
plt.figure(figsize=(12,5))

df['year_added'].value_counts().sort_index().plot()

plt.title("Content Added to Netflix by Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles Added")

plt.show()
```

# Key Insights

1. Movies significantly outnumber TV Shows on Netflix.

2. The United States contributes the largest amount of content to Netflix's catalog.

3. Drama and International Movies are among the most popular genres available on the platform.

4. Netflix experienced rapid growth in content additions after 2015.

5. Most Netflix titles are targeted toward mature audiences (TV-MA and TV-14 ratings).

6. A small number of directors contribute multiple titles, while most directors have only a few entries.

7. The majority of movies have moderate durations, with only a few extremely long titles.
